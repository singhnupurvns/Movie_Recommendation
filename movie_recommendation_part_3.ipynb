```python
from tmdbv3api import Movie
import requests
import numpy as np

tmdb_movie = Movie()

def get_genre(x):
    try:
        genres = []  # List to store genres
        result = tmdb_movie.search(x)  # Search for the movie
        if not result:  # Handle case where no results are found
            return np.nan

        movie_id = result[0].id  # Get the first movie's ID
        response = requests.get(
            'https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id, tmdb_movie.api_key)
        )

        data_json = response.json()  # Parse the JSON response
        if "genres" in data_json and data_json["genres"]:  # Check if genres exist
            for genre in data_json["genres"]:
                genres.append(genre["name"])  # Append genre names to the list
            return " ".join(genres)  # Return joined genre names as a string
        else:
            return np.nan  # Return NaN if no genres are available

    except Exception as e:
        print(f"An error occurred: {e}")
        return np.nan









```

```python
import numpy as np

# Define the get_actor3 function
def get_actor3(cast_and_crew):
    try:
        # Example logic to extract the third actor's name
        # Assuming 'Cast and crew' is a string like "Director: X, Actors: A, B, C"
        actors = cast_and_crew.split(", ")
        if len(actors) >= 3:
            return actors[2]  # Return the third actor
        else:
            return np.nan  # Return NaN if less than 3 actors are present
    except Exception as e:
        print(f"Error processing row: {cast_and_crew}. Error: {e}")
        return np.nan

# Apply the function to the column
df_2018["actor_3_name"] = df_2018["Cast and crew"].map(lambda x: get_actor3(x))
```
